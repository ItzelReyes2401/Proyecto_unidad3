# Marco Teórico

**Series de Tiempo**

Las series de tiempo son conjuntos de observaciones ordenadas cronológicamente, donde cada observación corresponde a un valor medido en un instante de tiempo. Estas series son fundamentales en áreas como la economía, finanzas, meteorología y muchas otras, ya que permiten analizar la evolución de variables a lo largo del tiempo. El análisis de series de tiempo busca identificar patrones, tendencias y relaciones dentro de los datos para hacer predicciones futuras.

**Promedio Móvil de Precios de Cierre**

El promedio móvil es una técnica estadística que se utiliza para suavizar una serie de tiempo. El promedio móvil de los precios de cierre es comúnmente usado en el análisis financiero, especialmente en mercados bursátiles, para identificar tendencias subyacentes al filtrar las fluctuaciones diarias. Existen distintos tipos de promedios móviles, siendo los más comunes el promedio móvil simple (SMA) y el promedio móvil exponencial (EMA). El SMA se calcula tomando el promedio de un número fijo de observaciones pasadas, mientras que el EMA da más peso a los valores más recientes. Esta herramienta es útil para determinar la dirección general de los precios y detectar puntos de inflexión en el mercado.

**Tendencia No Lineal a Precios de Cierre**

La tendencia no lineal en una serie de tiempo se refiere a un patrón de cambio que no puede ser descrito adecuadamente por una línea recta. A menudo, el comportamiento de las series temporales en mercados financieros sigue una trayectoria más compleja, que puede ser modelada utilizando polinomios de alto grado. Estos modelos pueden capturar fluctuaciones y cambios en la dinámica de los precios de una forma más precisa que los modelos lineales, permitiendo identificar no solo la dirección general del precio, sino también las aceleraciones y desaceleraciones en su comportamiento.

**Residuales de la Regresión Polinomial de Cierre**

Los residuales en una regresión polinómica son la diferencia entre los valores observados de la variable dependiente (en este caso, los precios de cierre) y los valores predichos por el modelo. Los residuales pueden indicar qué tan bien un modelo está ajustando los datos. Si los residuales son pequeños y aleatorios, el modelo puede considerarse adecuado. Sin embargo, grandes residuales podrían sugerir que el modelo no está capturando correctamente algunos aspectos de la serie temporal, lo que puede deberse a la presencia de patrones no modelados o a la influencia de factores externos.

**Componentes de Fourier de la Serie**

La Transformada de Fourier es una técnica matemática utilizada para descomponer una serie temporal en una suma de componentes sinusoidales (frecuencias) de diferentes longitudes de onda. Esta herramienta es útil para identificar patrones cíclicos y de alta frecuencia en los datos. En el análisis de series temporales, especialmente en finanzas, la Transformada de Fourier se utiliza para detectar ciclos recurrentes que podrían no ser evidentes a simple vista. La frecuencia dominante en la serie indica ciclos de mayor duración o importancia, mientras que las frecuencias menores representan fluctuaciones más rápidas.

**Prueba de Estacionariedad de Dickey-Fuller Aumentada**

La prueba de Dickey-Fuller aumentada (ADF) es una prueba estadística utilizada para determinar si una serie de tiempo es estacionaria o no. Una serie estacionaria es aquella cuya media, varianza y autocorrelación no cambian a lo largo del tiempo. En otras palabras, no presenta tendencias o patrones predecibles a largo plazo. La prueba ADF se basa en la hipótesis nula de que la serie tiene una raíz unitaria, lo que implica que no es estacionaria. Si el valor p obtenido es menor que un umbral de significancia (comúnmente 0.05), se rechaza la hipótesis nula, lo que implica que la serie es estacionaria.

**Prueba t Pareada para el Efecto Significativo de la Estacionalidad**

La prueba t pareada se utiliza para comparar dos series relacionadas, por ejemplo, los componentes de tendencia y estacionalidad de una serie de tiempo, para determinar si la estacionalidad tiene un efecto significativo en la serie. Esta prueba evalúa si las medias de dos muestras relacionadas son significativamente diferentes. En el contexto de las series de tiempo, se puede usar para comparar las diferencias entre los precios de cierre antes y después de eliminar los efectos estacionales, lo que ayuda a determinar si la estacionalidad tiene un impacto importante en el comportamiento de la serie.

**Gráficas de Descomposición**

La descomposición de series de tiempo es una técnica que separa una serie de tiempo en varios componentes: tendencia, estacionalidad y residuales. Las gráficas de descomposición permiten visualizar cómo cada uno de estos componentes contribuye al comportamiento general de la serie.

**Serie Real vs Serie Descompuesta**
 Muestra la serie original comparada con la suma de los componentes de tendencia, estacionalidad y residuales.
Serie Real vs Serie Descompuesta sin Residuales: Muestra solo la tendencia y la estacionalidad, eliminando el ruido de los residuales.
Serie Real vs Serie Descompuesta sin el Componente Estacional: Representa solo la tendencia y los residuales, permitiendo estudiar las fluctuaciones estructurales sin los patrones cíclicos.
Estas gráficas son útiles para comprender la dinámica interna de las series de tiempo, identificar patrones cíclicos y tendencias subyacentes.

**Causalidad de Granger**

La causalidad de Granger es un concepto utilizado para determinar si una serie temporal puede ayudar a predecir otra. La prueba de causalidad de Granger no implica causalidad en el sentido tradicional, sino que se refiere a la relación temporal entre dos series, es decir, si los valores pasados de una serie contienen información útil para predecir los valores futuros de otra serie. Si la prueba muestra un valor p bajo (menor que el nivel de significancia), se puede concluir que una serie "causa" a la otra en términos de predicción.

**Prophet y sus Predicciones**

Prophet es un modelo de predicción desarrollado por Facebook, diseñado para trabajar con series temporales que contienen fuertes efectos estacionales y festivos. Prophet es especialmente útil cuando las series de tiempo presentan tendencias no lineales y cambios de tendencia abruptos. Utiliza un enfoque aditivo en el que la serie se descompone en componentes de tendencia, estacionalidad y efectos de días festivos. Prophet permite realizar predicciones a corto y largo plazo y se ajusta fácilmente a las particularidades de las series temporales, lo que lo convierte en una herramienta robusta para hacer proyecciones precisas en contextos complejos.


El análisis de series de tiempo con herramientas como el promedio móvil, modelos polinómicos, la Transformada de Fourier, y pruebas estadísticas como la ADF y la causalidad de Granger, permite descomponer, entender y predecir el comportamiento de variables a lo largo del tiempo. Estas técnicas son fundamentales en contextos financieros, económicos y otros campos donde se busca identificar patrones, entender dinámicas subyacentes y realizar predicciones informadas. La utilización de modelos como Prophet complementa estos análisis al permitir predicciones adaptativas y flexibles que pueden mejorar la toma de decisiones en condiciones de incertidumbre.

```python
def Transformada_de_Fourier(serie, terminos, un_grafico):
  '''La función acepta los argumentos "serie", la cual debe de ser un array
  en numpy y en pandas de la forma df["my_variable"].values. Por otra parte,
  el parámetro "terminos" es un número natural que indica la cantidad de
  términos que desarrolla la serie. Entre mayor sea el número de términos,
  la serie será más precisa, pero más difícil de interpretar.

  La función tiene como salida una lista con los componentes sinosoidales de serie.
  El último elemento de la lista, es la suma de todos los componentes'''

  import numpy as np
  import matplotlib.pyplot as plt
  import pandas as pd
  from scipy.fftpack import fft, ifft

  # Supongamos que tienes una serie en df['AAPL_Close']
  # Reemplaza esta línea con tu DataFrame y serie específicos
  n = len(serie)

  # Calcular la transformada de Fourier
  transformada_fourier = fft(serie)

  # Obtener las frecuencias
  frecuencias = np.fft.fftfreq(n)

  # Crear un DataFrame para almacenar la frecuencia y su magnitud
  componentes = pd.DataFrame({
      'frecuencia': frecuencias,
      'magnitud': np.abs(transformada_fourier),
      'longitud_de_onda': 1 / frecuencias
  })

  # Ordenar el dataframe de mayor a menor en términos de magnitud

  componentes = componentes.sort_values(by='magnitud', ascending=False)

  # Seleccionar las frecuencias con mayor magnitud (excluyendo la frecuencia cero)
  top_frecuencias = componentes.loc[componentes['frecuencia'] > 0].nlargest(terminos, 'magnitud')
  top_frecuencias.reset_index(drop=True, inplace=True)
  print("Frecuencias principales:\n", top_frecuencias)

  # Crear el índice de tiempo para la serie
  t = np.arange(n)

  # Graficar cada componente de frecuencia junto con la serie original
  plt.figure(figsize=(12, 4))

  componente_temporal_sumado = np.zeros_like(serie)
  componentes_temporales = []

  n=0
  for i, row in top_frecuencias.iterrows():
      n+=1
      # Copiar la transformada de Fourier y mantener solo la frecuencia actual
      fourier_component = np.zeros_like(transformada_fourier)
      idx = np.where(frecuencias == row['frecuencia'])[0][0]  # índice de la frecuencia en la FFT
      fourier_component[idx] = transformada_fourier[idx]  # mantener solo la frecuencia positiva
      fourier_component[-idx] = transformada_fourier[-idx]  # mantener la frecuencia negativa correspondiente
      if n == (terminos+1):
        break

      # Reconstruir la señal en el tiempo
      componente_temporal = ifft(fourier_component).real
      componentes_temporales.append(componente_temporal)
      componente_temporal_sumado += componente_temporal

      # Graficar la componente
      plt.plot(
          componente_temporal,
          label=f'Longitud de onda {1 / row["frecuencia"]:.0f}',
          alpha=1,
          linewidth = 0.5,
          )
      plt.title('Componentes de Fourier de la Serie')
      plt.xlabel('Tiempo')
      plt.ylabel('Valor')
      plt.legend()
      plt.grid()

  componentes_temporales.append(componente_temporal_sumado)
  plt.plot(serie, label='Serie Original', color='black', alpha=0.5)
  if not un_grafico:
    plt.figure(figsize=(12, 4))
    plt.plot(serie, label='Serie Original', color='black', alpha=0.5)
  plt.plot(componente_temporal_sumado, label='Componente temporal sumada', color='red')
  plt.legend()
  plt.title('Suma de los Componentes de Fourier de la Serie')
  plt.xlabel('Tiempo')
  plt.ylabel('Valor')
  plt.grid()
  plt.show()

  return componentes_temporales
```

```python
# Descomponer la serie temporal
import statsmodels.api as sm
descomposicion = sm.tsa.seasonal_decompose(
    df['TSLA_Close'],
    model='additive',
    period=72,    # cada cuanto sube y baja el precio
    extrapolate_trend=0,
    two_sided=False
) # 252 días de negociación al año
fig = descomposicion.plot()
plt.show()
```

```python
#Prueba de estacionariedad: prueba de dickey-Fuller aumentada(ADF)

from statsmodels.tsa.stattools import adfuller

nivel_de_significancia = 0.05

#Realizar la prueba de Dickey-Fuller en la tendencia
adf_test = adfuller(descomposicion.resid.dropna())

print('Estadistadistico ADF:' , adf_test[0])
print('Valor-p:' , adf_test[1])
print('P-value', adf_test[1])

if adf_test[1] <= nivel_de_significancia:
    print('La tendencia es estacionaria (rechazamos la hipotesis nula)')
else:

    print('La tendencia no es estacionaria (aceptamos la hipotesis nula)')
```

```python
from scipy.stats import ttest_rel

# Prueba t pareada para el efecto significativo de la estacionalidad

#Hipótesis nula (H₀): Ambas series son iguales.
#Hipótesis alternativa (H₁): Ambas series son diferentes


nivel_de_significancia = 0.05
tendencia = descomposicion.trend
tendencia_estacionalidad = descomposicion.seasonal + descomposicion.trend

# Eliminar valores NaN de ambas series
tendencia.dropna(inplace=True)
tendencia_estacionalidad.dropna(inplace=True)

# Realizar la prueba t pareada
t_stat, p_valor = ttest_rel(tendencia, tendencia_estacionalidad)

print("Estadístico t:", t_stat)
print("Valor p:", p_valor)
print("\n")

# Interpretación de los resultados
if p_valor < nivel_de_significancia:
    print("El valor p es menor que 0.05, por lo tanto, rechazamos la hipótesis nula.")
    print("Conclusión: La estacionalidad tiene un efecto significativo en la serie de tiempo.")
else:
    print("El valor p es mayor o igual que 0.05, por lo tanto, no podemos rechazar la hipótesis nula.")
    print("Conclusión: La estacionalidad no tiene un efecto significativo en la serie de tiempo.")
```

```python
def Transformada_de_Fourier(serie, terminos, un_grafico):
  '''La función acepta los argumentos "serie", la cual debe de ser un array
  en numpy y en pandas de la forma df["my_variable"].values. Por otra parte,
  el parámetro "terminos" es un número natural que indica la cantidad de
  términos que desarrolla la serie. Entre mayor sea el número de términos,
  la serie será más precisa, pero más difícil de interpretar.

  La función tiene como salida una lista con los componentes sinosoidales de serie.
  El último elemento de la lista, es la suma de todos los componentes'''

  import numpy as np
  import matplotlib.pyplot as plt
  import pandas as pd
  from scipy.fftpack import fft, ifft

  # Supongamos que tienes una serie en df['AAPL_Close']
  # Reemplaza esta línea con tu DataFrame y serie específicos
  n = len(serie)

  # Calcular la transformada de Fourier
  transformada_fourier = fft(serie)

  # Obtener las frecuencias
  frecuencias = np.fft.fftfreq(n)

  # Crear un DataFrame para almacenar la frecuencia y su magnitud
  componentes = pd.DataFrame({
      'frecuencia': frecuencias,
      'magnitud': np.abs(transformada_fourier),
      'longitud_de_onda': 1 / frecuencias
  })

  # Ordenar el dataframe de mayor a menor en términos de magnitud

  componentes = componentes.sort_values(by='magnitud', ascending=False)

  # Seleccionar las frecuencias con mayor magnitud (excluyendo la frecuencia cero)
  top_frecuencias = componentes.loc[componentes['frecuencia'] > 0].nlargest(terminos, 'magnitud')
  top_frecuencias.reset_index(drop=True, inplace=True)
  print("Frecuencias principales:\n", top_frecuencias)

  # Crear el índice de tiempo para la serie
  t = np.arange(n)

  # Graficar cada componente de frecuencia junto con la serie original
  plt.figure(figsize=(12, 4))

  componente_temporal_sumado = np.zeros_like(serie)
  componentes_temporales = []

  n=0
  for i, row in top_frecuencias.iterrows():
      n+=1
      # Copiar la transformada de Fourier y mantener solo la frecuencia actual
      fourier_component = np.zeros_like(transformada_fourier)
      idx = np.where(frecuencias == row['frecuencia'])[0][0]  # índice de la frecuencia en la FFT
      fourier_component[idx] = transformada_fourier[idx]  # mantener solo la frecuencia positiva
      fourier_component[-idx] = transformada_fourier[-idx]  # mantener la frecuencia negativa correspondiente
      if n == (terminos+1):
        break

      # Reconstruir la señal en el tiempo
      componente_temporal = ifft(fourier_component).real
      componentes_temporales.append(componente_temporal)
      componente_temporal_sumado += componente_temporal

      # Graficar la componente
      plt.plot(
          componente_temporal,
          label=f'Longitud de onda {1 / row["frecuencia"]:.0f}',
          alpha=1,
          linewidth = 0.5,
          )
      plt.title('Componentes de Fourier de la Serie')
      plt.xlabel('Tiempo')
      plt.ylabel('Valor')
      plt.legend()
      plt.grid()

  componentes_temporales.append(componente_temporal_sumado)
  plt.plot(serie, label='Serie Original', color='black', alpha=0.5)
  if not un_grafico:
    plt.figure(figsize=(12, 4))
    plt.plot(serie, label='Serie Original', color='black', alpha=0.5)
  plt.plot(componente_temporal_sumado, label='Componente temporal sumada', color='red')
  plt.legend()
  plt.title('Suma de los Componentes de Fourier de la Serie')
  plt.xlabel('Tiempo')
  plt.ylabel('Valor')
  plt.grid()
  plt.show()

  return componentes_temporales
```

```python
# Descomponer la serie temporal
import statsmodels.api as sm
descomposicion = sm.tsa.seasonal_decompose(
    df['PCRFF_Close'],
    model='additive',
    period=125,    # cada cuanto sube y baja el precio
    extrapolate_trend=0,
    two_sided=False
) # 252 días de negociación al año
fig = descomposicion.plot()
plt.show()
```

```python
#Prueba de estacionariedad: prueba de dickey-Fuller aumentada(ADF)

from statsmodels.tsa.stattools import adfuller

nivel_de_significancia = 0.05

#Realizar la prueba de Dickey-Fuller en la tendencia
adf_test = adfuller(descomposicion.resid.dropna())

print('Estadistadistico ADF:' , adf_test[0])
print('Valor-p:' , adf_test[1])
print('P-value', adf_test[1])

if adf_test[1] <= nivel_de_significancia:
    print('La tendencia es estacionaria (rechazamos la hipotesis nula)')
else:

    print('La tendencia no es estacionaria (aceptamos la hipotesis nula)')
```

```python
from scipy.stats import ttest_rel

# Prueba t pareada para el efecto significativo de la estacionalidad

#Hipótesis nula (H₀): Ambas series son iguales.
#Hipótesis alternativa (H₁): Ambas series son diferentes


nivel_de_significancia = 0.05
tendencia = descomposicion.trend
tendencia_estacionalidad = descomposicion.seasonal + descomposicion.trend

# Eliminar valores NaN de ambas series
tendencia.dropna(inplace=True)
tendencia_estacionalidad.dropna(inplace=True)

# Realizar la prueba t pareada
t_stat, p_valor = ttest_rel(tendencia, tendencia_estacionalidad)

print("Estadístico t:", t_stat)
print("Valor p:", p_valor)
print("\n")

# Interpretación de los resultados
if p_valor < nivel_de_significancia:
    print("El valor p es menor que 0.05, por lo tanto, rechazamos la hipótesis nula.")
    print("Conclusión: La estacionalidad tiene un efecto significativo en la serie de tiempo.")
else:
    print("El valor p es mayor o igual que 0.05, por lo tanto, no podemos rechazar la hipótesis nula.")
    print("Conclusión: La estacionalidad no tiene un efecto significativo en la serie de tiempo.")
```

# CAUSALIDAD DE GRANGER


***Concepto y aplicación.***

La causalidad de Granger es un concepto estadístico desarrollado por Clive Granger en 1969, esta prueba permite determinar si una variable temporal ayuda a predecir otra. Dicho esto, la causalidad se define como una dependencia temporal, es decir, si los valores pasados de una variable contienen información útil para prever los valores futuros de otra.

La causalidad de Granger tiene diversas aplicaciones en diferentes disciplinas, en especial en aquellas que trabajan con series temporales.

1. **Economía y finanzas.**
Es ideal para la siguientes area en las que desarrollan la anteriores diciplinas.

  **Mercados financieros :**
- Evaluar si los movimientos en los precios del petróleo afectan los índices bursátiles.
- Estudiar la relación predictiva entre el tipo de cambio y los flujos de capital.
- Determinar si los volúmenes de negociación en la bolsa pueden predecir la volatilidad del mercado.

 **Macroeconomía :**

* Identificar si los cambios en los ingresos fiscales afectan el crecimiento del PIB.
* Analizar la relación entre la inversión extranjera directa y el desarrollo económico.

**Ciencias ambientales y climatologia.**
* Los fenómenos climáticos y ambientales dependen de interacciones complejas entre variables (temperatura, emisiones de CO₂, precipitaciones).
* Ejemplo: Determinar si el aumento de las emisiones de CO₂ "causa" un incremento en la temperatura global podría fortalecer la argumentación científica sobre el cambio climático.

**Medicina y Ciencias de la Salud**
* En salud, muchas variables son temporales (como niveles de glucosa, frecuencia cardíaca o propagación de enfermedades), y entender cómo interactúan es crucial para el diagnóstico y la prevención.
* Ejemplo: Si las alteraciones en la frecuencia cardíaca Granger-causan cambios en la presión arterial, los médicos podrían monitorear una variable para prever riesgos en la otra.

**Ciencias Sociales y Psicología.**
* En estas áreas, es común estudiar cómo los eventos sociales o psicológicos se influyen mutuamente a lo largo del tiempo.
* Ejemplo: Analizar si un aumento en los niveles de educación predice (causa) una disminución en las tasas de criminalidad.

**Ciencias Políticas.**
* Los eventos políticos (como protestas, cambios de gobierno o políticas públicas) tienen efectos temporales que necesitan análisis para la toma de decisiones estratégicas.
* Ejemplo: Determinar si el aumento de la violencia en una región Granger-causa un cambio en la popularidad de un líder político permite entender dinámicas electorales.

**Tecnología e Innovación.**
* Los sistemas tecnológicos producen grandes cantidades de datos temporales. Identificar relaciones causales permite optimizar sistemas y prevenir fallos.
* Ejemplo: En ciberseguridad, si ciertos patrones de tráfico de red Granger-causan ataques cibernéticos, se pueden establecer medidas preventivas.

Entre muchos otros campos de aplicación.

***Requisitos para aplicar la prueba.***

**Estacionariedad.** La prueba Granger requiere que las series temporales sean estacionarias, es decir, que las propiedades como la media, varianza y correlación sean constantes con el tiempo. De lo contrario es necesario transformarlas.

**Especificación del número de rezagos.** Para obtener un número de rezagos se utilizan los criterios de información como el criterio de Akaike (AIC) o el criterio Bayesiano de Schwarz (BIC). Esto asegura que el modelo sea parsimonioso pero captura la dinámica subyacente de las series.

**Datos suficientes.** La cantidad de datos debe de ser suficiente para obtener estimaciones confiables.

***Razones por las que se aplica.***

1. Toma de desiciones.
2. Versatilidad.
3. Interacción entre variables.
4. Complemento con otras técnicas.

```python
# prompt: Prueba de Granger

from statsmodels.tsa.stattools import grangercausalitytests

# Definir las series de tiempo que se van a analizar
serie1 = df['TSLA_Close']
serie2 = df['PCRFF_Close']

# Realizar la prueba de causalidad de Granger
resultados_granger = grangercausalitytests(df[['TSLA_Close', 'PCRFF_Close']], maxlag=60)

# Mostrar los resultados
for lag, resultado in resultados_granger.items():
  print(f'Lag {lag}:')
  for prueba, valor_p in resultado[0].items():
    print(f'Prueba {prueba}: Valor-p = {valor_p}')
  print()

# Analizar los valores-p para determinar si existe causalidad de Granger
nivel_de_significancia = 0.05
for lag, resultado in resultados_granger.items():
    for prueba, valor_p in resultado[0].items():
        # Access the p-value, which is the second element in the tuple
        p_value = valor_p[1]
        if p_value < nivel_de_significancia:
            print(f'Existe causalidad de Granger entre las series en lag {lag} para la prueba {prueba}')
        else:
            print(f'No existe causalidad de Granger entre las series en lag {lag} para la prueba {prueba}')

# Interpretación:
# - Si el valor-p es menor que el nivel de significancia (por ejemplo, 0.05),
#   se rechaza la hipótesis nula de que no existe causalidad de Granger.
#   Esto sugiere que la serie 2 causa la serie 1.
# - Si el valor-p es mayor que el nivel de significancia, no se rechaza la
#   hipótesis nula y no hay evidencia de causalidad de Granger.
```

**Estacionalidad:** Variaciones en tiempos determinados.

**Efectos puntuales o anomalías:** Eventos específicos que afectan los datos temporalmente.

* Utiliza tendencias lineales o algoritmos que pueden no ajustarse bien a series con comportamientos más complejos o no lineales.

# Resultados.

**Causalidad de Granger.**

La prueba de **causalidad de Granger** entre las series de precios de cierre ajustados de Tesla y Panasonic no mostró evidencia significativa de relación predictiva en ningún rezago (hasta 60 analizados). Los valores-p obtenidos en cada caso fueron mayores al nivel de significancia establecido (0.05), lo que indica que ninguna de las series contribuye a predecir de manera significativa los valores futuros de la otra. En otras palabras, la existencia de que exista una ralacion entre estas dos empresas es nula.

Esta ausencia de causalidad podría atribuirse a la falta de interdependencia directa entre las dos empresas en términos de mercado o datos no modelados. Es importante considerar que la prueba no establece causalidad real, sino únicamente relaciones predictivas basadas en datos históricos y que podría ser necesario incorporar factores adicionales para un análisis más completo.

**Prophet.**

El modelo Prophet aplicado a los precios de cierre ajustados de Tesla (TSLA) predijo una tendencia general creciente en el horizonte de 365 días proyectado para 2024. Este modelo identificó patrones estacionales moderados, como fluctuaciones anuales y semanales, pero sin cambios drásticos recientes. Las predicciones indican un crecimiento estable en los precios, con intervalos de confianza amplios hacia finales de 2024, reflejando una mayor incertidumbre en el largo plazo.

El ajuste del modelo fue adecuado para las tendencias históricas, capturando los comportamientos generales de los precios. Sin embargo, no se logro evitar fluctuaciones abruptas o eventos inesperados que pudieran afectar significativamente el mercado. Estos resultados sugieren que, aunque Prophet es útil para tendencias generales y estacionalidades, las predicciones dependerán también de factores externos y de dinámicas no incluidas en los datos históricos analizados.
